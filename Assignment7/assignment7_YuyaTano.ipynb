## 課題2

pandasでDataフォルダの```state-abbrevs.csv```、```state-areas.csv```、```state-population.csv```を読み込む。

これらのデータを使って、**2010年**アメリカの各州の人口密度を計算しよう。

````python
abbrevs = pd.read_csv(".\Data\state-abbrevs.csv")
areas = pd.read_csv(".\Data\state-areas.csv")
population = pd.read_csv(".\Data\state-population.csv")
df_try = pd.concat([abbrevs, areas, population], axis=1)
df_try[:51]
#全部アルファベット順にしてDCを上にもってくる。謎のプエルトリコを削除
#populationで2010年、totalを抽出
````

````python
abbrevs_sorted = abbrevs.sort_values("state").reset_index(drop=True)
#abbrevs_sorted = abbrevs_sorted.reset_index(drop=True)
abbrevs_sorted
````

````python
areas_sorted = areas.sort_values("state")
#areas_sorted = areas_sorted.reset_index(drop=True)
areas_sorted = areas_sorted[areas_sorted["state"] != "Puerto Rico"].reset_index(drop=True)
#areas_sorted
````

````python
#df.loc[(df["age"]>30)&(df["age"]<50)&(df["sex"]=="male"), ["name","age","sex"]]
pop_sorted = population.loc[(population["year"]==2010)&(population["ages"]=="total"), ["state/region","ages","year", "population"]]
pop_sorted = pop_sorted[:51]
#pop.sort_values("state/region").reset_index(drop=True) ←州名と略称のアルファベット順は違うので、sortしてはいけない
pop_sorted = pop_sorted.reset_index(drop=True)
pop_sorted
````

````python
df2 = pd.concat([abbrevs_sorted, areas_sorted, pop_sorted], axis=1)
df2
````

````python
df2["density"] = df2["population"] / df2["area (sq. mi)"]
df2
````

## 課題3

```state/region```、```ages```と```years```でデータフレームをグループして、各グループで$\frac{population}{area (sq. mi)}$の形で人口密度を計算してください。

````python
#mergeでabbrevs_sorted, areas_sorted とpopulationを結合
df3 = pd.merge(pd.merge(abbrevs_sorted, areas_sorted, on="state"), population, left_on="abbreviation", right_on="state/region")
#df3 = df3.drop("abbreviation", axis=1)
df3
````

````python
df3_group = df3.groupby(["state/region", "ages", "year"]).agg({"population": sum, "area (sq. mi)": "mean"})
df3_group["density"] = df3_group["population"] / df3_group["area (sq. mi)"]
df3_group
````
