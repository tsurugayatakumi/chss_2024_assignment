## 課題3

```state/region```、```ages```と```years```でデータフレームをグループして、各グループで$\frac{population}{area (sq. mi)}$の形で人口密度を計算してください。

````python
df7 = df2
df8 = df5["state/region"]
df7.loc[:,"state/region"] = df8
df9 = df7.set_index("state/region")
````

````python
df6 = df3.loc[(df3["state/region"] != "USA")]
df6_groupby = df6.groupby(["state/region","ages","year"], as_index=False).sum()
df10 = df6_groupby.set_index("state/region")
df10["population density"] = df10["population"] / df9["area (sq. mi)"]
df10
````
