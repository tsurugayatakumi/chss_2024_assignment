## 課題2

pandasでDataフォルダの```state-abbrevs.csv```、```state-areas.csv```、```state-population.csv```を読み込む。

これらのデータを使って、**2010年**アメリカの各州の人口密度を計算しよう。

````python
import pandas as pd

abbrevs_path = 'Data/state-abbrevs.csv'
areas_path = 'Data/state-areas.csv'
population_path = 'Data/state-population.csv'

abbrevs = pd.read_csv(abbrevs_path)
areas = pd.read_csv(areas_path)
population = pd.read_csv(population_path)

merged = pd.merge(population, abbrevs, how='outer', left_on='state/region', right_on='abbreviation')

merged = pd.merge(merged, areas, on='state')

data_2010 = merged[merged['year'] == 2010]

data_2010['density'] = data_2010['population'] / data_2010['area (sq. mi)']

result = data_2010[['state', 'density']].sort_values(by='density', ascending=False)

result
````

## 課題3

```state/region```、```ages```と```years```でデータフレームをグループして、各グループで$\frac{population}{area (sq. mi)}$の形で人口密度を計算してください。

````python
import pandas as pd

abbrevs_path = 'Data/state-abbrevs.csv'
areas_path = 'Data/state-areas.csv'
population_path = 'Data/state-population.csv'

abbrevs = pd.read_csv(abbrevs_path)
areas = pd.read_csv(areas_path)
population = pd.read_csv(population_path)

merged = pd.merge(population, abbrevs, how='outer', left_on='state/region', right_on='abbreviation')

merged = pd.merge(merged, areas, on='state')

def calculate_density(group):
    group['density'] = group['population'] / group['area (sq. mi)']
    return group

result = merged.groupby(['state/region', 'ages', 'year']).apply(calculate_density)

result = result[['state/region', 'ages', 'year', 'density']].sort_values(by=['state/region', 'ages', 'year'])

print(result)
````
