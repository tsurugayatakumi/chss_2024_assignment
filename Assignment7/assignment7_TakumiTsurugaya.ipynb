## 課題2

pandasでDataフォルダの```state-abbrevs.csv```、```state-areas.csv```、```state-population.csv```を読み込む。

これらのデータを使って、**2010年**アメリカの各州の人口密度を計算しよう。

````python
df_2=pd.read_csv("Data\state-abbrevs.csv")
df_3=pd.read_csv("Data\state-areas.csv")
df_4=pd.read_csv("Data\state-population.csv")
````

````python
df_2
````

````python
#dfの州表記を略称に
df_3["state"]=df_2["abbreviation"]
df_3
````

````python
#2010年と州全体人口で絞り込み
df_5=df_4[(df_4["year"]==2010)&(df_4["ages"]=='total')]
df_5
#列名をstateに変更
df_5.rename(columns={"state/region":"state"},inplace=True)
#面積データと人口データの合体
merged_df=pd.merge(df_3,df_5,on="state")
#州全体人口/面積
merged_df["density"]=merged_df["population"]/merged_df["area (sq. mi)"]
merged_df["density"]
#結果
merged_df
````

## 課題3

```state/region```、```ages```と```years```でデータフレームをグループして、各グループで$\frac{population}{area (sq. mi)}$の形で人口密度を計算してください。

````python
#state/regionの列名をstateに変更
df_4.rename(columns={"state/region":"state"},inplace=True)
#面積データと人口データの合体
merged_df_2=pd.merge(df_3,df_4,on="state")
#人口密度の定義
merged_df_2["density"]=merged_df_2["population"]/merged_df_2["area (sq. mi)"]
#グループ化＋無理やりdensityを合計することで結果を表示
gruop_df=merged_df_2.groupby(["state","ages","year"])["density"].sum()
gruop_df
````
