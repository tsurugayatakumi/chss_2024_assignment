## 課題2

pandasでDataフォルダの```state-abbrevs.csv```、```state-areas.csv```、```state-population.csv```を読み込む。

これらのデータを使って、**2010年**アメリカの各州の人口密度を計算しよう。

````python
abbrevs = pd.read_csv('Data\state-abbrevs.csv')
areas = pd.read_csv('Data\state-areas.csv')
population = pd.read_csv('Data\state-population.csv')

print(abbrevs.shape)
print(areas.shape)
print(population.shape)
````

````python
abbrevs.dropna()
areas.dropna()
population.dropna()

merged_df = pd.merge(abbrevs, areas, on='state')
display(merged_df.shape)
````

````python
merged_df = merged_df.rename(columns={'abbreviation':'state/region'})
````

````python
merged_df2 = pd.merge(merged_df, population, on= 'state/region')
display(merged_df2)
````

````python
merged_df3 = merged_df2.loc[(merged_df2["year"]==2010) & (merged_df2["ages"]=="total" )]
````

````python
merged_df3['density'] = merged_df3['population'] /merged_df3['area (sq. mi)']
````

````python
display(merged_df3)
````

## 課題3

```state/region```、```ages```と```years```でデータフレームをグループして、各グループで$\frac{population}{area (sq. mi)}$の形で人口密度を計算してください。

````python
merged_df4 = merged_df2.groupby(["state/region", "ages", "year"])
display(merged_df4.first())
````

````python
df5 = merged_df4.transform(lambda x:x)
#display(df5)
df5["density"] = df5.apply(lambda x: x["population"] / x["area (sq. mi)"], axis=1)
#print(type(df5))
display(df5)
````
