## 課題2

pandasでDataフォルダの```state-abbrevs.csv```、```state-areas.csv```、```state-population.csv```を読み込む。

これらのデータを使って、**2010年**アメリカの各州の人口密度を計算しよう。

````python
# read_csv関数はfile path(文字列の形式)をparamentとして受け取ります。
# file pathとしての文字列の先頭に「./」を使用すると、read_csv関数は現在のfolderからfileを検索し始めます。
state_abbrevs_df = pd.read_csv('./Data/state-abbrevs.csv')
state_areas_df = pd.read_csv('./Data/state-areas.csv')
state_population_df = pd.read_csv('./Data/state-population.csv')
````

````python
# state_abbrevsのcsvの中には、Puerto Ricoとその略語がないです。
# しかし、他のcsvの中には、Puerto Ricoに関する情報があるので、
# データ処理を容易にするために、Puerto Ricoとその略語をstate_abbrevsからのDataFrameに追加します。
state_abbrevs_df.loc[len(state_abbrevs_df)] = ['Puerto Rico', 'PR']
state_abbrevs_df = state_abbrevs_df.sort_values(by='state')
state_abbrevs_df = state_abbrevs_df.reset_index(drop=True)

state_areas_df = state_areas_df.sort_values(by='state')
state_areas_df = state_areas_df.reset_index(drop=True)
````

````python
# display_htmlを使用して2つのDataFrameを同じ行にdisplayします
from IPython.display import display_html
def display_side_by_side(*dfs):
    html_str = ''
    for df in dfs:
        html_str += df.to_html()
    display_html(html_str.replace('table', 'table style="display:inline"'), raw=True)
display_side_by_side(state_abbrevs_df, state_areas_df)
````

````python
# stateという列を使用してmergeします
combined_df = pd.merge(state_abbrevs_df, state_areas_df, on='state', how='outer')
combined_df
````

````python
population_2010_df = state_population_df[(state_population_df['ages'] == 'total') & (state_population_df['year'] == 2010)]
population_2010_df = population_2010_df.rename(columns={'state/region': 'abbreviation'})
population_2010_df= population_2010_df[population_2010_df['abbreviation'] != 'USA']
population_2010_df = population_2010_df.sort_values(by='abbreviation').reset_index(drop=True)
population_2010_df
````

````python
density_2010_df = pd.merge(combined_df, population_2010_df, on='abbreviation', how='outer')
density_2010_df['population density'] = density_2010_df['population'] / density_2010_df['area (sq. mi)']
density_2010_df = density_2010_df.sort_values(by='state').reset_index(drop=True)
density_2010_df
````

````python
# 厳密に言えば、Puerto Ricoは米国の州ではありません。
# これはPuerto Ricoに関するrowを削除されたDataFrameです。
without_PR_df = density_2010_df.drop(density_2010_df[density_2010_df['state'] == 'Puerto Rico'].index)
without_PR_df = without_PR_df.reset_index(drop=True)
without_PR_df
````

## 課題3

```state/region```、```ages```と```years```でデータフレームをグループして、各グループで$\frac{population}{area (sq. mi)}$の形で人口密度を計算してください。

````python
# mergeのため、combined_dfの列名をstateからstate/regionに変更します。
combined_df = combined_df.rename(columns={'state': 'state/region'})
density_by_group_df = state_population_df.rename(columns={'state/region': 'abbreviation'})
density_by_group_df.drop(density_by_group_df[density_by_group_df['abbreviation'] == 'USA'].index, inplace=True)
density_by_group_df = pd.merge(combined_df, density_by_group_df, on='abbreviation', how='outer')
density_by_group_df=density_by_group_df.groupby(['state/region','ages','year'])\
                                       .agg({'population':'sum','area (sq. mi)':'mean'}).reset_index()
density_by_group_df['population density'] = density_by_group_df['population'] / density_by_group_df['area (sq. mi)']
````
